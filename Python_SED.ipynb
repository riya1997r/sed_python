```python
from astropy.table import Table
from urllib.parse import quote

radius = 3
target = "10:27:32.07 +40:10:09.4"

url = f"https://vizier.cds.unistra.fr/viz-bin/sed?-c={quote(target)}&-c.rs={radius}"

sed = Table.read(url)
print(sed)
```

```python
import numpy as np
filters = sed['sed_filter']
print(np.array(filters))
```

```python
import numpy as np

print(np.array(np.unique(sed['sed_filter'])))
print(len(np.array(np.unique(sed['sed_filter']))))
```

```python
from astropy.table import unique

sed_unique = unique(sed, keys='sed_filter', keep='first')

print(len(sed_unique))
print(sed_unique)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('freq (GHz)')
ax.set_ylabel('Flux (Jsky)')
ax.scatter(sed_unique["sed_freq"], sed_unique["sed_flux"])
plt.yscale('log')
plt.xscale('log')
plt.xlim(1e4,1e7)
plt.show()
```

```python
filters = sed_unique['sed_filter']

# As a Python list
print(np.array(filters))
```

```python
selected_filters = ['GALEX:NUV',
    'SDSS:u',
    'SDSS:g',
    'SDSS:r',
    'SDSS:i',
    'SDSS:z',
    '2MASS:J',
    '2MASS:H',
    '2MASS:Ks',
    'WISE:W1',
    'WISE:W2',
    'WISE:W3',
    'WISE:W4'
]

mask = np.isin(sed_unique['sed_filter'], selected_filters)

sed_selected = sed_unique[mask]

print(sed_selected)
```

```python
sed_new = sed_unique[mask].copy()

print(sed_new)
```

```python
print(sed_new['sed_filter'].tolist())
plt.scatter(sed_new['sed_freq'], sed_new['sed_flux'])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Frequency (GHz)')
plt.ylabel('Flux (Jy)')
plt.show()
```

```python
print(sed_new.colnames)
```

```python
sed_small = sed_new[['sed_filter', 'sed_freq', 'sed_flux', 'sed_eflux']]

print(sed_small)
```

```python
sed_mJy = sed_small.copy()
sed_mJy['sed_flux'] *= 1000
sed_mJy['sed_eflux'] *= 1000

sed_mJy['sed_flux'].unit = 'mJy'
sed_mJy['sed_eflux'].unit = 'mJy'

print(sed_mJy)
```

```python
from astropy.table import Table
import numpy as np

source_id = "J102732.07+401009.4"   # replace with your source name as per SDSS
z = 0.12

sed_final = Table()

sed_final['source_id'] = [source_id] * len(sed_mJy)
sed_final['redshift'] = [z] * len(sed_mJy)
sed_final['flux_mJy'] = sed_mJy['sed_flux']
sed_final['flux_err_mJy'] = sed_mJy['sed_eflux']
sed_final['filter'] = sed_mJy['sed_filter']

print(sed_final)
```

```python
sed_final_mod_new_hand= sed_final.copy()

filter_map = {
"2MASS:H":'2mass.H',
'2MASS:J':'2mass.J',
'2MASS:Ks':'2mass.Ks',
'GALEX:NUV':'galex.NUV',
'SDSS:g':'sloan.sdss.g',
'SDSS:i':'sloan.sdss.i',
'SDSS:r':'sloan.sdss.r',
'SDSS:u':'sloan.sdss.u',
'SDSS:z':'sloan.sdss.z',
'WISE:W1':'wise.W1',
'WISE:W2':'wise.W2',
'WISE:W3':'wise.W3',
'WISE:W4':'wise.W4'
}

for i, f in enumerate(sed_final_mod_new_hand['filter']):
    if f in filter_map:
        sed_final_mod_new_hand['filter'][i] = filter_map[f]

print(sed_final_mod_new_hand)
```
